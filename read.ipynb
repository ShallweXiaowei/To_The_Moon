```python
import pandas as pd
import json
from pprint import pprint


with open("CIK0000320193.json", 'r') as file:
    data = json.load(file)


facts = data['facts']


def read_item(data, key):
    try:
        return pd.DataFrame(data["facts"]["us-gaap"][key]["units"]["USD"])
    except:
        return pd.DataFrame(data["facts"]["us-gaap"][key]["units"]["USD/shares"])
```
